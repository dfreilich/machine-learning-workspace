## Part I. Understanding of Measures of Influence

### Influence-Directed Explanations:
---
1. What is a quantity of interest, what is a distribution of interest? Please provide a brief definition   
  **Answer**: The _distribution_ and _quantity_ of interest together capture the aspects of the network that we're trying to explain. The _distribution of interest_ is the input (or distribution, from which the input came) which is fed to the network – whether that be a single instance (gradient at a point), a collection of cat images, the distribution of all images, or the difference between an image and a baseline. The _quantity of interest_ is the machine's 'understanding'/behavior for which we are trying to find explanations – whether that be understanding why a particular image is classified as a cat, how it distinguishes cats from dogs, or more.  


2. In the query “what is the influence of neuron 0 toward the cat class?”, what is the distribution of interest and quantity of interest?  
 **Answer**: The _distribution of interest_ is neuron 0 (the object you are trying to analyze), and the _quantity of interest_ is 'influence towards the cat class' (the behavior the network displayed).  
 
 
3. In the query “what is the influence of pixel (5,0) toward a cat image C, compared to a dog image D?”, what is the distribution of interest and quantity of interest?  
 **Answer**: The _distribution of interest_ is pixel (5,0) (the object you are trying to analyze), and the _quantity of interest_ is the comparative outcome towards 'cat', versus 'dog'.

```python
##Defining Helper Functions
#Given an array of activations, returns the top neurons, the top values for those neurons, and the indices for that neuron
def get_max_values(arr, num):
    #Gets max values for each neuron into (# neurons, ) array
    im_max_values = np.amax(np.amax(arr, axis=2), axis=1)
    #Returns a list of the top $(num) values, ordered
    top_neurons = im_max_values.argsort()[-num:][::-1]
    top_vals = im_max_values[top_neurons]
    top_coords = []
    for i in range(0, num):
        # Gets the full coordinates of the highest value
        coords = (np.unravel_index(np.argmax(arr[top_neurons[i]]), arr[top_neurons[i]].shape))
        new = [top_neurons[i]] + list(coords)
        top_coords.append(new)
    return top_neurons, top_vals, top_coords

#Get Attributions for a single Neuron, explaining an image
def get_attrib_for_neuron(neuron, im):
    Q = model_imnet.layers[19].input[:, neuron, :, :].max()
    expl = Explanation(model_imnet).for_input().for_quantity(Q)
    attribs = expl.explain(im, ignore_difference_term=True)
    return attribs

#Given a list of neurons and possibly ims, gets attribs for all of them and visualizes in a tiler
def visualize_neurons(neuron_list, im_list):
    attribs = []
    if len(im_list) > 1:
        for i in range (0, len(neuron_list)) :
            attribs.append(get_attrib_for_neuron(neuron_list[i], im_list[i]))
    else:
        for neuron in neuron_list:
            attribs.append(get_attrib_for_neuron(neuron, im_list[0]))
    visualizer_tiler = VisualizerTiler(AttributionMask())
    plt.figure()
    plt.title('Visualizing ' + str(len(neuron_list)) + ' most influential neurons');
    output = visualizer_tiler.visualize(attribs,[0,255]);
    for im in im_list:
        plt.figure()
        plt.imshow(np.array(im.transpose(1,2,0),dtype = 'int'))
```

### How does the explanations of using internal neurons compared with the explanations of integrated gradients?
Using internal neurons gives you a different flavor for what is important to the network. The integrated gradients told you that the most important parts of the image, for the decision, were the starfish parts, which makes sense; but the internal neurons let you zero in on what individual neurons learned – in this case, that the edge between a starfish limb and the ground was critical for seeing that there is a starfish there.

## PART IV. Model Compression

```python
import seaborn as sns; 
sns.set()

## Function to get attributions for a class and average them
def get_av_attrs(cl, ims):
    l_19_explainer = Explanation(model_imnet).for_layer(19).absolute(cl)
    l_19_attrs = []
    for i in range(0, ims.shape[0]):
        l_19_attrs.append(l_19_explainer.explain(ims[i], ignore_difference_term=True).attribution)
    print(len(l_19_attrs))
    print(l_19_attrs[0].shape)
    cl_av_attrs = np.average(l_19_attrs, axis=0)
    return cl_av_attrs

## Function to take an array of attributions and return a list of alpha (top) and beta (bottom) values from it
def get_alpha_beta_vals(arr, alpha, beta):
    #Gets max values for each neuron into (# neurons, ) array
    attr_max_values = np.amax(np.amax(arr, axis=2), axis=1)
    #Returns a list of the top alpha values and bottom beta values
    combined_vals = []
    alpha_vals = []
    # Fix for a bug, where alpha = 0 would return the entire list
    if(alpha > 0):
        alpha_vals = np.argpartition(attr_max_values, -alpha)[-alpha::]
        combined_vals.extend(alpha_vals)
    beta_vals = np.argpartition(attr_max_values, beta)[:beta]
    combined_vals.extend(beta_vals)
    return alpha_vals, beta_vals, combined_vals

## Function to compress a model, given a list of combined alpha and beta values to preserve
def compress(alpha_beta_vals):
    model = models.vgg_16_imagenet()
    weights = model.layers[18].get_weights()
    new_weights = np.zeros(weights[0].shape)
    new_bias = np.zeros(weights[1].shape)
    for val in alpha_beta_vals:
        new_weights[: , : , : ,val] = weights[0][:,:,:,val]
        new_bias[val] = weights[1][val]
    model.layers[18].set_weights((new_weights, new_bias))
    return model

## Function to compute the recall of a model
def compute_recall(model, ims, label):
    preds = model.predict(ims)
    class_preds = np.argmax(preds, axis=1)
    label_preds = sum(class_preds==label)
    return float(label_preds)/ims.shape[0]

## Function to evaluate how an alpha (num) and beta (num) value will work, on a set of attributions
def eval_alpha_beta(attrs, alpha, beta, ims, cl):
    _, _, combined_vals = get_alpha_beta_vals(attrs, alpha, beta)
    compressed_model = compress(combined_vals)
    rec = compute_recall(compressed_model, ims, cl)
    return compressed_model, rec 

## Function to cycle through the alpha and beta range, orchestrated by their step sizes, 
## and find the best recall, together with the associated recall_value map, alpha, beta, and model
def get_best_compression(alpha_range, alpha_step, beta_range, beta_step, attrs, ims, cl):
    best_recall = 0
    best_alpha = 0
    best_beta = 0
    best_model = None
    recall_vals = []
    x_labels = []
    y_labels = []
    for alpha in range(0, alpha_range, alpha_step):
        x_labels.append(alpha)
        alpha_row = []
        for beta in range(0, beta_range, beta_step):
            ## Want to ensure that it doesn't keep on appending beta every single cycle
            if len(y_labels) <= ((beta_range/beta_step)):
                y_labels.append(beta)
            model, rec = eval_alpha_beta(attrs, alpha, beta, ims, cl)
            alpha_row.append(rec)
            if(rec > best_recall):
                best_alpha = alpha
                best_beta = beta
                best_recall = rec
                best_model = model
        recall_vals.append(alpha_row)
    return [recall_vals, x_labels, y_labels], best_alpha, best_beta, best_recall, best_model

## Function to visualize the recall values on a heatmap
def visualize_heatmap(plot_vals):
    plt.figure(figsize=(10,10))
    ax = sns.heatmap(np.array(plot_vals[0]).T, vmin=0, vmax=1, annot=True, xticklabels=plot_vals[1], 
                     yticklabels=plot_vals[2], cbar_kws={"orientation": "horizontal"})
    plt.title('Compression Heat Map')
    plt.xlabel('Alpha Values')
    plt.ylabel('Beta Values')
```

```python
##Define Class Numbers
chainsaw_class = 491
bonnet_class= 452
park_bench_class = 703
sloth_bear_class = 297
pelican_class = 144

## Load Class images
chainsaw_ims = test_data.get_chainsaws()
bonnet_ims = test_data.get_bonnets()
park_bench_ims = test_data.get_park_benches()
sloth_bear_ims = test_data.get_sloth_bears()
pelican_ims = test_data.get_pelicans()

## Get Average Class Attributions
cs_av_attrs = get_av_attrs(chainsaw_class, chainsaw_ims)
bon_av_attrs = get_av_attrs(bonnet_class, bonnet_ims)
pb_av_attrs = get_av_attrs(park_bench_class, park_bench_ims)
sb_av_attrs = get_av_attrs(sloth_bear_class, sloth_bear_ims)
pel_av_attrs = get_av_attrs(pelican_class, pelican_ims)
```

```python
## Get base recall for classes
chainsaw_base_recall = compute_recall(model_imnet, chainsaw_ims, chainsaw_class)
bonnet_base_recall = compute_recall(model_imnet, bonnet_ims, bonnet_class)
park_benches_recall = compute_recall(model_imnet, park_benches_ims, park_bench_class)
sloth_bears_recall = compute_recall(model_imnet, sloth_bears_ims, sloth_bear_class)
pelicans_recall = compute_recall(model_imnet, pelicans_ims, pelican_class)

print(chainsaw_base_recall)
print(bonnet_base_recall)
print(park_benches_recall)
print(sloth_bears_recall)
print(pelicans_recall)
```
