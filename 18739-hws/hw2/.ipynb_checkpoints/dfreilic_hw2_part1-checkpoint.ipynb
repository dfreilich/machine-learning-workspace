```python
import gzip
import pickle as pkl
import numpy as np
path = 'mnist.pkl.gz'
def load_data(path):
    f = gzip.open(path, 'rb')
    try:
        #Python3
        train_set, valid_set, test_set = pkl.load(f, encoding='latin1')
    except:
        #Python2
        train_set, valid_set, test_set = pkl.load(f)
    f.close()
    return(train_set,valid_set,test_set)
train_set,valid_set,test_set = load_data(path)
X_train,Y_train = train_set
X_test,Y_test = test_set
X_val, Y_val = valid_set
print(X_train.shape,Y_train.shape)
```

## Upgrade I 

```python
#HW1 model, please revise the model
import keras
from keras.utils import np_utils
from keras.models import Sequential 
from keras.layers import Dense, Activation 
from keras import initializers, optimizers
import time

#define variables
input_dim = 784 
nb_classes = 10
output_dim = nb_classes = 10 
batch_size = 64 
epoch = 50
l_r = 0.01
rho = 0.99
variance_scaling = initializers.VarianceScaling()

#Transform Y's to categorical
Y_train = np_utils.to_categorical(Y_train, nb_classes) 
Y_test = np_utils.to_categorical(Y_test, nb_classes)
Y_val = np_utils.to_categorical(Y_val, nb_classes)
print(X_train.shape,Y_train.shape)

def basic_model():
    model = Sequential() 
    model.add(Dense(64,input_dim=input_dim, kernel_initializer=variance_scaling, bias_initializer='zeros', activation='relu'))
    model.add(Dense(10, kernel_initializer=variance_scaling, bias_initializer='zeros', activation="softmax"))
    return model
def run_model(model, X_train, X_val, X_test):
    rmsprop = optimizers.RMSprop(lr=l_r, rho=rho)
    model.compile(optimizer=rmsprop, loss='categorical_crossentropy', metrics=['accuracy']) 
    start_time = time.time()
    model.fit(X_train, Y_train, batch_size=batch_size, epochs=epoch,verbose=1, validation_data=(X_val, Y_val)) 
    end_time = time.time()
    print(model.summary())
    print('Run Time for Training:', end_time-start_time)
    score = model.evaluate(X_test, Y_test) 
    print('Test accuracy:', score[1])
basic_model = basic_model()
run_model(basic_model, X_train, X_val, X_test)
```

* Test Accuracy: 0.9665
* Run Time: 104.376
* Number of Epochs: 50
* Batch Size: 64

## Upgrade II

```python
from keras.layers import Dropout, Flatten, Conv2D, MaxPooling2D
from matplotlib import pyplot as plt
%matplotlib inline

#Reshape the x-train data
X_train = np.reshape(X_train, ([X_train.shape[0],28,28,1]))
print(X_train.shape)
X_val = np.reshape(X_val, ([X_val.shape[0],28,28,1]))
X_test = np.reshape(X_test, ([X_test.shape[0],28,28,1]))

#redefine variables
epoch = 5
l_r = 0.001

# plt.imshow(X_train[100].squeeze())

# define the CNN model
def cnn_model():
    model = Sequential()
    model.add(Conv2D(32, (5,5), activation='relu', input_shape=(28,28,1), kernel_initializer=variance_scaling))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.20))
    model.add(Flatten())
    model.add(Dense(128, kernel_initializer=variance_scaling, bias_initializer='zeros', activation='relu'))
    model.add(Dense(10, kernel_initializer=variance_scaling, bias_initializer='zeros', activation='softmax'))
    return model


cnn_model = cnn_model()
run_model(cnn_model, X_train, X_val, X_test)
```

* Test Accuracy: 0.985
* Run Time: 103.9
* Number of Epochs: 5
* Batch Size: 64
* Learning Rate: 0.001

2) If an image I has dimension (width, height, depth) of W × H × D and is passed through a convolutional layer with N filters of filter size F, stride S and zero padding Z, what is the dimension of the filtered output for I?

    Answer: ((W+2*Z-F)/S) + 1, ((H+2*Z-F)/S) + 1, F

3) In this model, what is the dimension of the output of each neuron of the second layer(Maxpooling) for a batch of images (size 28× 28× N)?

    Answer: 12*12

```python
dropout = 0.20
epoch = 15

def my_model():
    model = Sequential()
    model.add(Conv2D(32, (5,5), activation='relu', input_shape=(28,28,1), kernel_initializer=variance_scaling))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(dropout))

    model.add(Conv2D(32, (3,3), activation='relu', kernel_initializer=variance_scaling))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(dropout))

    model.add(Conv2D(32, (3,3), activation='relu', kernel_initializer=variance_scaling))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(dropout))

    model.add(Flatten())
    model.add(Dense(128, kernel_initializer=variance_scaling, bias_initializer='zeros', activation='relu'))
    model.add(Dropout(dropout))
    model.add(Dense(10, kernel_initializer=variance_scaling, bias_initializer='zeros', activation='softmax'))

    return model
my_model = my_model()
run_model(my_model, X_train, X_val, X_test)
```

* Test Accuracy: 0.99
* Run Time: 466.17
* Number of Epochs: 15
* Batch Size: 64
* Learning Rate: 0.001
